```python
import numpy as np
import paho.mqtt.client as mqtt
import matplotlib.animation as animation
import time

## Define function to print messages received
def on_message(client, userdata, message):
    time.sleep(1)
    print("message received " ,str(message.payload.decode("utf-8")))
    # print("message topic=",message.topic)
    # print("message qos=",message.qos)
    # print("message retain flag=",message.retain)

    
    
## Create MQTT client to read & publish messages
# broker_address = "192.168.200.5"                     # [CHANGE IP ADDRESS HERE] --> IP address doesn't work because each device has different IP address
client = mqtt.Client()                                 # Create the client
client.connect("broker.mqttdashboard.com")       # Connect to MQTT broker on HiveMQ Broker


## Bind function to callback
client.on_message = on_message                         # Bind function to callback


## Start loop
client.loop_start()


## Subscribe to topic
print("subscribing...")
client.subscribe("testtopic/topic_maxg378")
time.sleep(0.5)                                        # The time delay (in seconds) is critical in giving enough time to publish data


## Publish message to topic
print("publishing...")
for i in np.arange(5):
    tempString = "hiadsfasdfasdfassdfafa " + str(i)
    client.publish("testtopic/topic_maxg378", tempString)                  # Publish (topic name, message)
    time.sleep(1)                                        # The time delay (in seconds) is critical in giving enough time to publish data


## Stop loop
client.disconnect()
client.loop_stop()
print("done")


```
